## HOSPITAL IMAGING: BASIC WORKFLOW CONCEPTS<a id="concepts"></a>

### What do I mean by “imaging” ?
* Diagnostic Imaging
  * Obtained to elicit a differential diagnosis or confirm a clinical suspicion
* Procedural Imaging
  * Obtained before, during, and after surgical and percutaneous invasive procedures
* Evidence Imaging
  * Captured primarily for documentation of a patient’s current state
* Image-Based Clinical Reports
  * i.e., image-enriched textual reports where thumbnail image depictions of lesions identified

### Radiology AND Enterprise Imaging
Defined as “a set of strategies, initiatives and workflows implemented across a healthcare enterprise to consistently and optimally capture, index, manage, store, distribute, view, exchange, and analyze all clinical imaging and multimedia content to enhance the electronic health record.”

Source: A Foundation for Enterprise Imaging: HIMSS-SIIM Collaborative White Paper
http://link.springer.com/article/10.1007/s10278-016-9882-0
Christopher J. Roth, Louis M. Lannum, Kenneth R. Persons

### Worth 1,000 Words
“While written descriptions of lesions and infectious  processes may be detailed,  the visual presentation of  these conditions is much more accurate and effective for diagnosis and treatment.”
Imaging: New Electronic Tool For Clinicians, 2003 (Susan I. Reynolds)

### ACCESS TO IMAGES IS SERIOUS BUSINESS: If I Can’t Find It, It Does Not Exist
If imaging studies are not readily available, patients may need to be re-imaged
* Increased patient risk (radiation, hospital-borne infections)
* Increased cost to the patient / healthcare system (unnecessary exams)
* Increased patient and caregiver discomfort and anxiety
* Increased delay for the patient (while re-imaging takes place)
* Increased delay for other patients (while re-imaging takes place)
* Or need for increased unnecessary capacity

### Simple Imaging Dept Workflow
1. Patient admitted to clinic
2. Doctor creates order
3. Order gets reviewed
4. Images get acauired
5. Report is written
6. Results are distributed

### Basic imaging workflow
1. Modality acquires images and transmits them to archive
2. Stores the images to a PACS (Picture Archive and Communication System)
3. Radiologist requests images for reading
4. Radiologist creates the report

### Data Representation in Healthcare
The Complete Health Record brings together discrete data, non-discrete data, and imaging.

Discrete data: HL7
* HL7 v2.x : 1989
* HL7 v3 : 2005
* FHIR : 2014

Non-discrete data
* CDA : 2000
* XDS : 2007
* Other formats like PDF and TIFF

Imaging data: DICOM
* DICOM : 1993
* WADO : 2003
* DICOMweb : ~2011

### What Technologies Enable Interoperability?
* DICOM® and DICOMweb™
  * Used for imaging-centric activities
    * E.g., PACS, post-processing - to retrieve, process, and parse the DICOM instances
  * Retrieval of imaging
  * Raw files with metadata as well as various renderings
  * Both a transport and a structure specification, and integration (Application Hosting)
* HL7® and FHIR®
  * Used for information-centric activities
    * E.g., EMRs, EHRs - to discover clinical data and work with order- and report-centric metadata
  * Retrieval of all patient record data (e.g., demographics)
  * Discrete results data (labs) and non-discrete (reports)
  * Both a transport and a structure specification, and integration (SMART on FHIR)

### Who’s in the Data Ecosystem?
* PACS and VNA
  * Image archive systems that are DICOM compliant
  * Store and retrieve are the most basic of use cases
* EMR and ancillary systems
  * Databases of health data that are HL7 compliant
  * Almost has all data necessary to create holistic view
* Health Records and IT
  * Custodians for all of the health data and passwords
  * Keys to authenticating to various systems as well as signing on your users

### Flavors of PACS
* All-in-one / Reconstructed
  * “Picture Archive Communication System” 
  * All-in-one solutions 
* Deconstructed
  * Can be broken apart into its composite pieces
	* Picture = Viewer software
	* Archive = VNA software
	* Communication = Workflow software
  * Buy components and integrate
* Hospitals may have legacy systems, they may have separate systems for every –ology (radiology, cardiology, dermatology, pathology, etc etc)
* Some solutions are pure web (in a browser), web-based (delivered using web technology with a thin client), a thick client, and/or mobile variants
* Solutions may be integrated into other hospital solutions, like EMRs
